# 1. Cleaning

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# demographic dataset from: https://www.census.gov/data/tables/time-series/demo/popest/2010s-counties-detail.html
# election dataset from: https://github.com/tonmcg/US_County_Level_Election_Results_08-20/blob/master/2016_US_County_Level_Presidential_Results.csv

# race data is a 2015 estimate
# election data is from the 2016 presidential election
```

```python
race_df = pd.read_csv('countypopmonthasrh.csv')
race_df.columns
```

```python
race_df = race_df[race_df['agegrp'] == 0]
race_df = race_df[race_df['yearref'] == 8]
race_df = race_df.fillna(0)
race_df.drop(['sumlev', 'sumlevstr','yearref','vintage','year','agegrp'], axis=1, inplace=True)
```

```python
races_with_sex = race_df.columns[7:]
races_with_sex
```

```python
races_clean = set()
for i in races_with_sex:
    r = i.split('_')
    races_clean.add(r[0])
        
races_clean = list(races_clean)
```

```python
for race in races_clean:
    race_col = []
    for row in range(len(race_df)):
        total = race_df[race+'_male'].iloc[row] + race_df[race+'_female'].iloc[row]
        race_col.append(total)
    race_df[race+'_total'] = race_col
    race_df = race_df.drop([race+'_male', race+'_female'], axis=1)
    
race_df.head()
```

```python
df_sum = race_df[race_df.columns[8:]].sum().to_frame().reset_index()
df_sum.columns = ['grp', 'pop']
df_sum.plot.bar(x='grp',y='pop')
plt.show()
```

```python
election_df = pd.read_csv('2016_US_County_Level_Presidential_Results.csv', index_col=0)
election_df['county'] = election_df['combined_fips']
election_df.drop(['combined_fips'], axis=1, inplace=True)
election_df.head()
```

```python
import random
random_county = random.choice(range(len(race_df)))
county_id = race_df.iloc[random_county]['county']
print(election_df[election_df['county'] == county_id])
print(race_df.iloc[random_county][:4])
```

```python
# Creating a dataset for reference
race_df.drop(['ctyname', 'state', 'stname'], axis=1, inplace=True)
election_df.drop(['state_abbr', 'county_name'], axis=1, inplace=True)
cty_ref = pd.read_csv('countypopmonthasrh.csv')
cty_ref = cty_ref[cty_ref['yearref'] == 8]
cty_ref = cty_ref[cty_ref['agegrp'] == 0][['county','ctyname', 'stname']].reset_index().drop('index', axis=1)
cty_ref.head()
```

```python
# changing each race into a proportion of the total population rather than total number
for race in races_clean:
    race_df[race+'_prop'] = race_df[race+'_total'] / race_df['tot_pop']
    race_df.drop([race+'_total'], axis=1, inplace=True)
    
race_df.head()
```

```python
final_df = pd.merge(race_df, election_df)
final_df.head()
```

# 2. EDA and some modeling

#### Which race has the strongest linear correlation with the percent dem votes?

```python
from sklearn.linear_model import LinearRegression
```

```python
def dem_lin_reg_race(race):
    reg = LinearRegression()
    X = np.array(final_df[race+'_prop']).reshape(-1,1)
    y = final_df['per_dem']
    reg.fit(X, y)
    return [reg, X, y]
```

```python
dem_corr = []
for race in races_clean:
    r = dem_lin_reg_race(race)
    dem_corr.append(r[0].score(r[1], r[2]))
    print(race, r[0].score(r[1], r[2]))
```

```python
def dem_reg_viz(race):
    reg, X, y = dem_lin_reg_race(race)
    plt.scatter(X*100, y*100, color='b')
    plt.plot(X*100, reg.predict(X)*100, color='r', linewidth=3)
    plt.ylabel('Vote Recieved by Democratic Party (%)', fontsize =10)
```

```python
dem_reg_viz('wa')
plt.xlabel('Pop. Identifying as White Alone (%)', fontsize=10)
plt.show()
```

```python
dem_reg_viz('ba')
plt.xlabel('Pop. Identifying as Black Alone (%)', fontsize=10)
plt.show()
```

```python
dem_reg_viz('aa')
plt.xlabel('Pop. Identifying as Asian Alone (%)', fontsize=10)
plt.ylim(0,100)
plt.show()
```

These three major race identification groups (White, Black, and Asian alone) had the highest r^2 values for their linear regression models, though still having weak correlations. Note: I didn't include the "combined" race groups like "alone OR in combination" or Hispanic/non-Hispanic that also had high r^2 values because they typically had very similar values to their just "alone" counterparts.

I think it's important to note that for the White Alone and Black Alone visualizations especially, it's apparent that a lower correlation may be caused not by more outliers and scattered data, but by a huge "block" of counties with close to zero or 100 with widely varying proportion of votes recieved by the Democratic party.

### Multiple Linear Regression

Obviously, race and elections are unlikely to be explained by a singular linear relationship. Let's see how all of the races groups collectively correlate with the dem vote.

```python
r = LinearRegression()
X = final_df[[x+'_prop' for x in races_clean]]
y = final_df['per_dem']
r.fit(X,y)
r.score(X,y)
```

Using all of the groups results in a fairly good model - with a higher r^2 value than any single-group regression we tried before. Additionally, our residual plot is (kind of) scattered and shows no pattern - a good indicator that a linear model is the way to go. 

EDA takeaways:
- found the race grps which have the highest correlation of determination with dem vote
- some correlations are higher/lower than with gop vote, overall consistent however
- using all groups results in a better correlation than any other singular group
- strong inverse relationship between black alone and white alone proportions
- some groups, like the indigenous, have extremely skewed data and not a lot of counties for each bin
- way more counties leaned gop than dem

## Modeling

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(final_df, stratify=final_df['winner'], test_size= .3, random_state=1)
```

```python
count = train['winner'].value_counts()
count = count.to_frame().reset_index()
count['party'] = count['index']
count['num wins'] = count['winner']
count.drop(['index', 'winner'], axis=1, inplace=True)
count.plot.bar('party', 'num wins')
plt.show()
```

```python
count = test['winner'].value_counts()
count = count.to_frame().reset_index()
count['party'] = count['index']
count['num wins'] = count['winner']
count.drop(['index', 'winner'], axis=1, inplace=True)
count.plot.bar('party', 'num wins')
plt.show()
```

```python
X_train = train[[x+'_prop' for x in races_clean]]
X_test = test[[x+'_prop' for x in races_clean]]
y_train = train['per_dem']
y_test = test['per_dem']
```

#### What's the best combination of races for predicting the winner? 

```python
best_combo = []
best_corr = 0
for race in races_clean:
    combo = set()
    combo.add(race)
    X=X_train[[r+'_prop' for r in combo]]
    y=y_train
    reg = LinearRegression().fit(X,y)
    if reg.score(X,y) > best_corr:
        best_corr = reg.score(X,y)
        best_combo = list(combo)
    for add_race in races_clean:
        combo.add(add_race)
        X=X_train[[r+'_prop' for r in combo]]
        y=y_train
        reg = LinearRegression().fit(X,y)
        if reg.score(X,y) > best_corr:
            best_corr = reg.score(X,y)
            best_combo = list(combo)

print(best_combo)
print(best_corr)
```

```python
len(races_clean) == len(best_combo)
```

```python
for race in races_clean:
    if race not in best_combo:
        print(race)
```

```python
X = X_train[[r+'_prop' for r in best_combo]]
reg = LinearRegression().fit(X, y_train)
```

```python
y_pred = reg.predict(X)
resid = y_train - y_pred
sns.regplot(x=y_pred,y=resid)
plt.xlabel('Predicted Prop. of Votes for Dem. Party')
plt.ylabel('Residual')
plt.ylim(-0.5,0.5)
plt.title('Residual Plot for the MLR Model - Training Data')
plt.show()
```

```python
# How well does it predict a random county's prop?
rand = random.choice(range(len(X_test)))
x = np.array(X_test[[r+'_prop' for r in best_combo]].iloc[rand]).reshape(1,-1)
print("Prediction: ", reg.predict(x))
print("Actual: ", y_test.iloc[rand])
```

The model is alright. It doesn't make outrageously incorrect predictions, but it sometimes makes predictions that change the outcome of the race in that county, which is not a good thing for accurately predicting the winner in that county.

## So, is census data on race an accurate predictor of a county's election results?

To an extent. 

In this notebook I explored the relationships between census race categories and election results in the respective counties. Some race categories are fairly positively correlated with the proportion of the vote won by one county, others negatively. Description as Hispanic/Non-hispanic was very weakly correlated with the result. Racial demographics are complex - two counties may have low black populations, but one may be majority white and the other majority indigenous, and their respective results would be rather different. The relationship between race and politics is extremely complex, and elections are influenced by various other factors as well. This analysis serves as a baseline for what factors may influence elections. Also, it would be interesting to see whether including only the eligible voting population makes for a stronger correlation.
